1. Exiba um gráfico do tipo box splot a partir dos seguintes dados:



salarios = [6900,7500,4700,11997,22000,16550,9655,8670,15090,29000,7600,14980,1250,1400,2600,7240,3250]

```python
import matplotlib.pyplot as plt
```

```python
salarios = [6900,7500,4700,11997,22000,16550,9655,8670,15090,29000,7600,14980,1250,1400,2600,7240,3250]
```

```python
plt.boxplot(salarios)
plt.title('Boxplot dos Salários')
plt.ylabel('Salário')
plt.show()
```

Uma outra opção:

```python

plt.boxplot(salarios, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot de Salários')
plt.ylabel('Salário')
plt.grid(True)
plt.show()
```

2. A partir dos dados a seguir, reproduza o stackplot abaixo:


dias = [1,2,3,4,5,6]

horas_dormindo = [6,7,5,8,6,7]

horas_familia = [2,2,1,2,1,1]

horas_trabalhando = [5,7,10,8,6,9]

horas_exercitando = [3,3,0,1,3,2]

```python
dias = [1,2,3,4,5,6]

horas_dormindo = [6,7,5,8,6,7]

horas_familia = [2,2,1,2,1,1]

horas_trabalhando = [5,7,10,8,6,9]

horas_exercitando = [3,3,0,1,3,2]
```

```python
plt.stackplot(dias, horas_dormindo, horas_familia, horas_trabalhando, horas_exercitando, labels=['Dormindo','Família','Trabalhando','Exercitando'], colors=['green', 'blue', 'red', 'orange'])
plt.title('Exemplo de rotina de 6 dias')
plt.xlabel('Dias')
plt.ylabel('Atividade')
plt.legend(loc='lower right')
plt.show()
```

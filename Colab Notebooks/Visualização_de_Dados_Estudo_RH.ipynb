```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Mineração dos Dados


```python
 employee_df = pd.read_csv('/content/Human_Resources.csv')
```

```python
#Transformando ‘yes’, ‘no’ em zeros e uns.
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x: 1 if x == 'Yes'else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x: 1 if x == 'Yes'else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x == 'Y' else 0)
employee_df.head()
```

```python
#Remover atributos que não representam tanta informação.
employee_df.drop(['EmployeeCount', 'StandardHours',
'Over18', 'EmployeeNumber'], axis = 1, inplace=True)
```

```python
#Iremos agora dividir nosso dataset em pessoas que saíram da empresa (Attrition == 1) e pessoas que não saíram da empresa.
left_df = employee_df[employee_df['Attrition'] == 1]
stayed_df = employee_df[employee_df['Attrition'] == 0]
```

```python
#Verificar os que saíram da empresa
left_df.describe()
#Verificar os que ficaram na empresa
stayed_df.describe()
```

Machine Learning

```python
# Algoritmos de aprendizado de máquina, não trabalham com atributos categóricos (Texto). Precisamos transformá-los em atributos numéricos.
# Vamos criar um dataframe temporário:
X_cat = employee_df[['BusinessTravel', 'Department',
'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat
```

```python
#Adequando os atributos de acordo com a quantidade de classes:
# Male Female
# 1 0
# 0 1
# Single Married Divorced
# 1 0 0
# 0 1 0
# 0 0 1
```

```python
#Adequando os atributos de acordo com a quantidade de classes:
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
#Observar o número de colunas, agora 26:
X_cat.shape
#ransformando em dataframe novamente
type(X_cat)
X_cat = pd.DataFrame(X_cat)
type(X_cat)
employee_df['BusinessTravel'].unique()
```

```python
#Adequando os atributos numéricos:
X_numerical = employee_df[['Age', 'DailyRate',
'DistanceFromHome', 'Education', 'EnvironmentSatisfaction',
'HourlyRate', 'JobInvolvement', 'JobLevel',
'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate',
'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
'PerformanceRating', 'RelationshipSatisfaction',
'StockOptionLevel', 'TotalWorkingYears'
,'TrainingTimesLastYear' , 'WorkLifeBalance',
'YearsAtCompany' ,'YearsInCurrentRole',
'YearsSinceLastPromotion', 'YearsWithCurrManager']]
X_numerical
```

```python
#Unindo os dois dataframes:
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
#Agora vamos deixá-los na mesma escala, isto é, vamos normalizá-los:
X_all.columns = X_all.columns.astype(str)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
X
```

```python
# Vamos criar um dataframe com a variável que queremos prever,separadamente
y = employee_df['Attrition']
y
```

```python
#Dividindo a base entre treino e teste:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size = 0.3)
X_train.shape, y_train
```

Modelagem Preditiva

```python
#Importando bibliotecas:
from sklearn.linear_model import LogisticRegression
#Nome do nosso modelo:
logistic = LogisticRegression()
#Treinamento do algoritmo (fit: encaixar os dados). Apenas a base de dados de treinamento é utilizada
logistic.fit(X_train, y_train)
```

```python
#Realizando previsões. X_test são registros que o algoritmo nunca viu.
y_pred = logistic.predict(X_test)
y_pred
```

```python
# Observando as KPI’s: acurácia
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Observando as KPI’s: matriz de confusão
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
sns.heatmap(cm, annot=True, fmt='d');
```

```python
# Visualização dos Dados
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
precision_score(y_test, y_pred)
recall_score(y_test, y_pred)
f1_score(y_test, y_pred, average='macro')
print(classification_report(y_test, y_pred))
```

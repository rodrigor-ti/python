```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
agrupador = KMeans(n_clusters=4)
agrupador.fit(X)
```

```python
y_agrupador = agrupador.predict(X)
```

Visualização do resultado do K-Means:

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y_agrupador, s=50, cmap='viridis')

centers = agrupador.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9);
```

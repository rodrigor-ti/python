producao = [300, 450, 500, 400]

Com base nos seguintes dados:
1. Crie um gráfico de barras com os dados fornecidos.
2. Adicione um título ao gráfico: "Produção das Fábricas em 2023".
3. Adicione rótulos aos eixos X e Y: "Fábricas" e "Produção (unidades)".
4. Adicione uma cor diferente para cada barra.

```python
producao = [300, 450, 500, 400]
fabricas = ['Fábrica A', 'Fábrica B', 'Fábrica C', 'Fábrica D']
```

```python
cores = ['red', 'blue', 'green', 'orange']
plt.bar(fabricas, producao, color=cores)
plt.title('Produção das Fábricas em 2023')
plt.xlabel('Fábricas')
plt.ylabel('Produção (unidades)')
plt.show()
```

Exercício 1.3
temperatura = [22, 23, 21, 25, 24, 26, 27, 28, 29, 30]
umidade = [60, 55, 65, 50, 52, 48, 45, 43, 40, 38]
Com base nos seguintes dados:
Reproduza o seguinte gráfico em Python:
Dica: utilize mapa de cores (cmap) da
biblioteca matplotlib.

```python
temperatura = [22, 23, 21, 25, 24, 26, 27, 28, 29, 30]
umidade = [60, 55, 65, 50, 52, 48, 45, 43, 40, 38]
```

```python
scatter = plt.scatter(temperatura, umidade, c=umidade, cmap='viridis')

plt.title('Relação entre Temperatura e Umidade')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Umidade (%)')

cbar = plt.colorbar(scatter)
cbar.set_label('Umidade (%)')
plt.show()
```

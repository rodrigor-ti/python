```python
def converte_minusculo(texto):
    return texto.lower()

def remove_espacos(texto):
    return texto.strip()

def remove_numeros(texto):
    return ''.join([char for char in texto if not char.isdigit()])

def limpa_lista_textos(lista_textos):
    return [
        remove_numeros(
            remove_espacos(
                converte_minusculo(texto)
            )
        )
        for texto in lista_textos

    ]
```

```python
frases = ['Essas frases cont3m palavras que estÃ£o inc0rretas e precisam ser el1mind4s',
          'Bem d3 longe dava para ouvir os gritos da torc;da',
          'No 38.Âº disco de sua carreira Elba combina releituras e algumas canc@es inÃ©ditas'
]

lista_limpa = limpa_lista_textos(frases)
for texto in lista_limpa:
        print("Lista Limpa: ", texto)
```

```python
def criar_lista():
    lista = []
    while True:
        item = input("Informe um item para adicionar Ã  lista (ou digite 'sair' para finalizar): ")
        if item.lower() == 'sair':
            break
        lista.append(item)
    return lista
```

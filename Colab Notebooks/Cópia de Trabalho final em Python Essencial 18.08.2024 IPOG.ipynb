Escreva uma função e dê o nome **limpa_lista_textos** que chama 3 funções:


1.   converte_minusculo
2.   remove_espacos
3.   remove_numeros

Procure usar List Comprehension quando precisar percorrer a lista.

```python
def converte_minusculo(texto):
    return texto.lower()

def remove_espacos(texto):
    return texto.strip()

def remove_numeros(texto):
    return ''.join([char for char in texto if not char.isdigit()])

def limpa_lista_textos(lista_textos):
    return [
        remove_numeros(
            remove_espacos(
                converte_minusculo(texto)
            )
        )
        for texto in lista_textos
    ]
```

```python
lista_teste= ["  Python é Vida!!!", "Python é Legal 100%  ", "30 Alunos aprenderam Python em 18.08   ", "18/08/2024 Domingo   ", "PYTHON ESSENCIAL",]
lista_limpa = limpa_lista_textos(lista_teste)
print("Lista Limpa: ", lista_limpa)
```

Otimizando com função para criarmos a lista e depois chamar e depois tratar.

```python
def criar_lista():
    lista = []
    while True:
        item = input("Informe um item para adicionar à lista (ou digite 'sair' para finalizar): ")
        if item.lower() == 'sair':
            break
        lista.append(item)
    return lista
```

```python
lista_limpa = limpa_lista_textos(criar_lista())
print("Lista Limpa: ", lista_limpa)
```

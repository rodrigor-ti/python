```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Mineração dos Dados

```python
#Verificar se existem atributos nulos
sns.heatmap(employee_df.isnull(), cbar = False);

```

```python
# Visualização 1: Distribuição de funcionários por departamento
plt.figure(figsize=(10,6))
sns.countplot(data=employee_df, x='department', order=employee_df['department'].value_counts().index)
plt.title('Distribuição de Funcionários por Departamento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualização 2: Distribuição de salário por nível de cargo
plt.figure(figsize=(10,6))
sns.boxplot(data=employee_df, x='job_level', y='salary')
plt.title('Distribuição de Salário por Nível de Cargo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualização 3: Relação entre tempo de casa e salário
plt.figure(figsize=(10,6))
sns.scatterplot(data=employee_df, x='tenure', y='salary', hue='department')
plt.title('Tempo de Casa (em meses) x Salário')
plt.tight_layout()
plt.show()
```

```python
#Visualização 4: Distribuição de salário por gênero e nível de cargo
plt.figure(figsize=(12,6))
sns.boxplot(data=employee_df, x='job_level', y='salary', hue='gender')
plt.title('Distribuição Salarial por Gênero e Nível de Cargo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Visualização 5: Taxa de rotatividade (turnover) por departamento

employee_df['turnover'] = employee_df['active_status'].apply(lambda x: 'Ativo' if x == 1 else 'Desligado')

turnover_by_dept = employee_df.groupby(['department', 'turnover']).size().unstack(fill_value=0)
turnover_by_dept['taxa_desligamento'] = turnover_by_dept['Desligado'] / (turnover_by_dept['Desligado'] + turnover_by_dept['Ativo'])

turnover_by_dept = turnover_by_dept.sort_values('taxa_desligamento', ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=turnover_by_dept.index, y=turnover_by_dept['taxa_desligamento'])
plt.title('Taxa de Desligamento por Departamento')
plt.ylabel('Taxa de desligamento (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualização 6: Funcionários ativos por cidade e tipo de vínculo

ativos_df = employee_df[employee_df['active_status'] == 1]

pivot_city = ativos_df.pivot_table(index='location_city', columns='employment_status', aggfunc='size', fill_value=0)

pivot_city.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Funcionários Ativos por Cidade e Tipo de Vínculo')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de Funcionários')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusão

✅ Análise de Dados de RH (employees.csv)
A análise exploratória com base nas seis visualizações geradas permitiu identificar pontos importantes que podem contribuir diretamente para a tomada de decisões estratégicas na área de Recursos Humanos da empresa. Abaixo, seguem os principais destaques:

🔹 1. Distribuição de Funcionários por Departamento
Percebe-se uma concentração significativa de colaboradores em determinados departamentos, enquanto outros têm participação mínima. Essa diferença pode sinalizar uma possível desigualdade na alocação de pessoas ou ainda prioridades distintas entre áreas. Esses dados ajudam a orientar futuras decisões sobre contratações e eventuais reestruturações.

🔹 2. Salário por Nível de Cargo
A relação entre cargo e salário segue um padrão esperado: cargos mais altos tendem a receber mais. Entretanto, a presença de outliers chama atenção e pode indicar exceções ou distorções que merecem uma análise mais criteriosa, seja por erro, seja por situações específicas.

🔹 3. Tempo de Casa (Tenure) x Salário
Existe uma correlação entre o tempo de casa e o salário, mas ela não é uniforme. Em alguns casos, profissionais com longo tempo na empresa continuam com salários baixos, o que pode ser um indício de falta de políticas claras de crescimento ou de valorização por tempo de serviço.

🔹 4. Salário por Gênero e Cargo
Foram observadas diferenças salariais entre homens e mulheres em alguns níveis de cargo. Embora diversos fatores possam influenciar isso (como experiência ou tempo na empresa), esse achado reforça a necessidade de atenção à equidade salarial e à igualdade de oportunidades.

🔹 5. Rotatividade por Departamento
Certos departamentos apresentaram níveis de desligamento acima da média. Isso pode estar ligado ao clima de trabalho, estilo de liderança ou até mesmo falta de perspectiva de crescimento. Esses dados ajudam a direcionar ações voltadas à retenção de talentos e melhoria do ambiente organizacional.

🔹 6. Funcionários Ativos por Cidade e Tipo de Vínculo
A distribuição dos vínculos contratuais varia bastante conforme a cidade, sugerindo práticas regionais distintas. Isso pode abrir espaço para analisar a viabilidade de expandir modelos de trabalho mais flexíveis, como o remoto ou o híbrido, em locais onde eles já vêm sendo aplicados com sucesso.

🧠 Recomendações Estratégicas
Com base nos padrões observados, algumas ações recomendadas são:

Reavaliar as políticas de promoção e progressão salarial para garantir transparência e meritocracia;

Monitorar e reduzir possíveis disparidades salariais entre gêneros;

Focar em departamentos com maior rotatividade, buscando entender e resolver causas internas;

Considerar a expansão de modelos de trabalho mais flexíveis, especialmente onde já existe boa aceitação;

Investigar os casos que fogem ao padrão (outliers), pois podem representar tanto problemas quanto boas práticas.

## Outras visualizações no processo de Analise Exploratoria:

```python
# Mapeamento ordinal para os níveis de cargo
job_level_map = {
    'Individual Contributor': 1,
    'Team Lead': 2,
    'Manager': 3,
    'Director': 4,
    'Executive': 5
}


employee_df['job_level_num'] = employee_df['job_level'].map(job_level_map)

num_cols = ['salary', 'tenure', 'job_level_num']
corr = employee_df[num_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de Correlação entre Salário, Tempo de Casa e Nível de Cargo')
plt.tight_layout()
plt.show()
```

```python
# Criar faixas de tempo de casa
employee_df['faixa_tenure'] = pd.cut(employee_df['tenure'], bins=[0, 12, 36, 60, 120, np.inf],
                                      labels=['<1 ano', '1-3 anos', '3-5 anos', '5-10 anos', '10+ anos'])

pir_data = employee_df.groupby(['faixa_tenure', 'gender']).size().unstack().fillna(0)
pir_data['Feminino'] = -pir_data.get('Female', 0)
pir_data['Masculino'] = pir_data.get('Male', 0)

plt.figure(figsize=(10,6))
plt.barh(pir_data.index, pir_data['Feminino'], color='lightpink', label='Feminino')
plt.barh(pir_data.index, pir_data['Masculino'], color='lightblue', label='Masculino')
plt.title('Pirâmide de Tempo de Casa por Gênero')
plt.xlabel('Número de Funcionários')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Contar número de liderados por cada nível gerencial
levels = ['first_level_manager', 'second_level_manager', 'third_level_manager', 'fourth_level_manager']
liderados = {}

for level in levels:
    counts = employee_df[level].value_counts()
    liderados[level] = counts

liderados_df = pd.DataFrame(liderados).fillna(0).astype(int).T

plt.figure(figsize=(10,6))
sns.heatmap(liderados_df, cmap='YlGnBu', annot=True, fmt="d")
plt.title('Matriz de Liderança: Quantidade de Liderados por Nível Gerencial')
plt.xlabel('ID do Gestor')
plt.ylabel('Nível Gerencial')
plt.tight_layout()
plt.show()
```

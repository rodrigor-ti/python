1 e 10, caso um número inválido seja digitado, avise o usuário e peça o número
novamente; caso um número válido seja digitado, imprima o número.

**Primeiro Jogo em Python**

Gere um número aleatório entre 1 e 9 (incluindo 1 e 9) usando random do Python. Peça ao usuário para adivinhar o número e diga se ele adivinhou muito baixo, muito alto ou exatamente certo. (Dica: lembre-se de usar input() para receber o número do usuário)
Continue o jogo até que o usuário digitar 0
Mostre quantos chutes o usuário precisou para acertar


```python
import random
# random - número entre 0 e 1
numero_aleatorio = random.randint(1,9)
tentativas = 0
numero_chutado = 0

while numero_chutado != numero_aleatorio :
  numero_chutado = int(input("Adivinhe um número entre 1 e 9 (ou digite 0 para sair): "))
  tentativas += 1
  if numero_chutado == 0:
    print("Você saiu do jogo.")
    break
  elif numero_chutado < numero_aleatorio:
      print("Você chutou muito baixo!!!")
  elif numero_chutado > numero_aleatorio:
      print("Você chutou muito alto!!!")
  else:
      print(f"Parabéns! Você acertou o número em {tentativas} tentativas.")




```

**Parte 2 – Containers (coleções)**

1 - Criar uma lista de compras (entre 5 e 10 itens);
2 - Ordenar em ordem alfabética a lista;
3 - Apresentar a quantidade de itens da lista;
4 - Apresentar a lista toda.

```python
lista_compras = ["arroz", "feijão", "ovo", "leite", "pão", "café"]


while true:
  # Solicita ao usuário para digitar um número
    item= input("Digite um número (ou 0 para parar): ")
    if item == 0:
        break  # Sai do loop se o usuário digitar 0
    lista_compras.append("manteiga")
    print(lista_compras)
    break



lista_compras.sort()
print("Lista de Compras Ordenada: ", lista_compras)
print("Quantidade de itens da Lista de Compras: ",len(lista_compras))
```

**Exercício: listas em Python**

Explicação do exercício: https://www.youtube.com/watch?v=SXJKAVVlvGA

Crie um programa que lê vários números e os coloca em uma lista.
 Depois disso, mostre:

A. Quantos números foram digitados;

B. A lista de valores, ordenada de forma decrescente;

C. Se o valor 5 está ou não presente na lista.

```python
# Inicializa a lista que vai armazenar os números
numeros = []

while True:
    # Solicita ao usuário para digitar um número
    numero = int(input("Digite um número (ou 0 para parar): "))

    if numero == 0:
        break  # Sai do loop se o usuário digitar 0


    numeros.append(numero)  # Adiciona o número à lista

# Ordena a lista em ordem decrescente
numeros.sort(reverse=True)

print("-="* 30)2
# Exibe a quantidade de números digitados
print(f"\nVocê digitou {len(numeros)} número(s)!!!")

# Exibe a lista de valores em ordem decrescente
print(f"Lista de valores ordenada de forma decrescente: {numeros}")

# Verifica se o valor 5 está presente na lista
if 5 in numeros:
    print("O valor 5 está presente na lista!")
else:
    print("O valor 5 não está presente na lista!")
```

**Crie um dicionário com o nome e data de nascimento de 6 celebridades:
Milton Nascimento (26/10/1942), Anitta (30/03/1993), Neymar(05/02/1992), Pelé (23/10/1940), Djavan (27/01/1949) e Renato Russo (27/03/1960).**

Informe o usuário de quais celebridades seu programa conhece a data de nascimento.

Pergunte de quem o usuário gostaria de saber a data de nascimento.
Caso ele digite um nome válido, informe a data de nascimento daquela pessoa.

Caso ele digite um nome inválido, informe que seu programa não sabe a data de nascimento daquela pessoa.

```python
# Dicionário com nomes e datas de nascimento de 6 celebridades
celebridades = {
    "Milton Nascimento": "26/10/1942",
    "Anitta": "30/03/1993",
    "Neymar": "05/02/1992",
    "Pelé": "23/10/1940",
    "Djavan": "27/01/1949",
    "Renato Russo": "27/03/1960"
}

# Informa ao usuário quais celebridades o programa conhece
print("Eu conheço a data de nascimento das seguintes celebridades:")
for nome in celebridades.keys():
    print(nome)

# Solicita ao usuário o nome de uma celebridade para consultar a data de nascimento
nome_consulta = input("\nDigite o nome de quem você gostaria de saber a data de nascimento: ")

# Verifica se o nome digitado está no dicionário e informa a data de nascimento ou uma mensagem de erro
if nome_consulta in celebridades:
    print(f"{nome_consulta} nasceu em {celebridades[nome_consulta]}.")
else:
    print(f"Desculpe, eu não sei a data de nascimento de {nome_consulta}.")
```

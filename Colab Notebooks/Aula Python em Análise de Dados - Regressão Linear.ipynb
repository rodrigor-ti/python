```python
import pandas as pd
```

Dados de financeiros

```python

money_data_frame = pd.read_csv("https://drive.google.com/uc?id=1FtG5iX48UGSi7V8CTf_luw7T8GCsu8kI")
money_data_frame.head()
```

Quanto as pessoas s√£o felizes?!

```python
happiness_data_frame = pd.read_csv("https://drive.google.com/uc?id=1A4PzcQLIM6i2JSmp3WPAbEDUaxMAfMJf")
happiness_data_frame.head(40)
```

```python
money_happiness_df = pd.merge(happiness_data_frame, money_data_frame,
                              on='Country')
```

```python
money_happiness_df.head(40)
```

Salario!!!
Argentina
13588.84

```python
money_happiness_df.plot(kind = "scatter", x = "GDP", y = "Happiness")

```

```python
from sklearn.linear_model import LinearRegression

X = money_happiness_df [["GDP"]]
y = money_happiness_df [["Happiness"]]

```

```python

modelo = LinearRegression()
modelo.fit(X , y)
```

```python
import matplotlib.pyplot as plt
prediction_y = modelo.predict(X)
plt.scatter(X, y,  color='blue')
plt.plot(X,prediction_y, color='red')
plt.show()
```

```python
#Argentina 13588.84
# Filtrando o DataFrame para o pa√≠s "Argentina" fazendo uma predi√ß√£o seria talves: [[5.94150885]]
argentina_gdp = [[13588.84]]

argentina_happiness = modelo.predict(argentina_gdp)
print(argentina_happiness)
```
